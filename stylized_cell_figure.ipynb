```python
from stylized_module.stylized_cell import Stylized_Cell
from neuron import h
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load geometry into pandas dataframe
geo_standard = pd.read_csv('geom_standard.csv',index_col='id')

cell = Stylized_Cell(geometry=geo_standard)
```

```python
cell.all
```

```python
def get_pts(cell,idx):
    x = [cell.all[idx].x3d(0),cell.all[idx].x3d(1)]
    y = [cell.all[idx].y3d(0),cell.all[idx].y3d(1)]
    return (x,y)
```

```python
%matplotlib notebook

lw = 10
labels=['basal','trunk','tuft','soma']
fig = plt.figure(figsize=(5,6))
for i in range(6,8):
    label = labels[2] if i==6 else None
    x,y = get_pts(cell,i)
    plt.plot(x,y,color='purple',linewidth=lw,label=label)

x,y = get_pts(cell,5)
plt.plot(x,y,color='pink',linewidth=lw,label=labels[1])

for i in range(6,8):
    x,y = get_pts(cell,i)
    plt.plot(x,y,color='purple',linewidth=lw)

for i in range(1,5):
    label = labels[0] if i==1 else None
    x,y = get_pts(cell,i)
    plt.plot(x,y,color='b',linewidth=lw,label=label)

plt.plot([0,0],[0,0],linestyle="None",marker='.',color='r',markersize=50,label=labels[3])
plt.xlim([-240,400])
plt.ylim([-200,600])
ax = plt.gca()
ax.legend(loc=5,fontsize=20,frameon=False)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_linewidth(2)
ax.spines['bottom'].set_linewidth(2)
xtk = list(range(-200,400,200))
ytk = list(range(0,600,250))
ax.set_xticks(xtk)
ax.set_xticklabels([str(x) for x in xtk],fontsize=25)
ax.set_yticks(ytk)
ax.set_yticklabels([str(x) for x in ytk],fontsize=25)
ax.tick_params(width=3)
plt.xlabel('x',fontsize=40)
plt.ylabel('y',fontsize=40)
plt.show()
```
